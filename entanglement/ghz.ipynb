 ### A GHZ (Greenberger-Horne-Zeilinger) state is like a bell state but for 3 or more qubits—
 ### It’s an extremely strong entanglement where all qubits are perfectly correlated.

 

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

```python
# create a quantum circuit with 3 qubits

qc = QuantumCircuit(3)

# Apply hadamard gate to the first qubit
qc.h(0)

# Apply CNOT to qubits 2 and 3 with the first qubit as control
qc.cx(0, 1)
qc.cx(0, 2)
```

```python
# measure the qubits
qc.measure_all()

# simulate and run the circuit
simulator = AerSimulator()
result = simulator.run(qc).result()
counts = result.get_counts()


plot_histogram(counts) # the output must have |000> and |111> only
```
