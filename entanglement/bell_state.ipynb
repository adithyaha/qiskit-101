```python
from qiskit import QuantumCircuit, transpile
from qiskit.visualization import plot_histogram
from qiskit_aer import AerSimulator




# for additional visualization
from qiskit.visualization import plot_state_city
from qiskit.quantum_info import DensityMatrix
```

```python
# Create a 2-qubit quantum circuit
qc = QuantumCircuit(2)

# Create Entanglement: Hadamard + CNOT
qc.h(0)
# qc.x(1) # if you want to switch up between |00>|11> and |01>|10>
qc.cx(0, 1)

```

```python
# Simulate the circuit 
# (without measurement for visualization of state)
simulator = AerSimulator()


 # Save quantum state for visualisation
qc.save_statevector()

result = simulator.run(qc).result()
statevector = result.get_statevector()

# get density matrix of the state
dm = DensityMatrix(statevector)

# visualize
plot_state_city(dm)
```

```python
# Measure both qubits - once measured, state collapses
qc.measure_all()
counts = result.get_counts()
```

```python

# if entanglement works- only two results: 00 and 11
plot_histogram(counts)
```
