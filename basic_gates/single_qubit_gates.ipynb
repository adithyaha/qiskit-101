```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector
from qiskit_aer import AerSimulator
```

```python
# create a single-qubit quantum circuit
qc = QuantumCircuit(1) 

# Display the empty circuit
qc.draw("mpl")
```

```python
# The Hadamard gate: Puts the qubit into superposition
# If the initial state is |0>: H|0> -> 1/2^1/2(|0> + |1>)
# If the initial state is |1>: H|1> -> 1/2^1/2(|0> - |1>)

qc.h(0)
qc.draw("mpl")
```

```python
# Pauli Gates: X, Y, Z

# qc.x(0) # X (NOT) gate - flips |0> to |1> or vice-versa
# qc.y(0) # Y gate - applies a \pi rotation around the y-axis
# qc.z(0) # z gate - applies a \pi phase shift

qc.draw('mpl')
```

```python
# Phase gates: S and T
# qc.s(0)  # S gate (π/2 phase shift)
# qc.t(0)  # T gate (π/4 phase shift)
# qc.draw('mpl')
```

```python
# Simulate the circuit and visualize

simulator = AerSimulator()
qc.save_statevector() # Save the state vector for visualization
result = simulator.run(qc).result()
statevector = result.get_statevector()

# Plot the Bloch sphere representation
plot_bloch_multivector(statevector)
```
